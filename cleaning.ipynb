```python
import pandas as pd
```

```python
bios = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Olympics-Dataset/master/athletes/bios.csv')
```

```python
bios.head()
```

```python
df = bios.copy()
```

```python
# Fixing the name
df['name'] = df['Used name'].str.replace('•', ' ')
```

```python
# Spliting the measurements of the athletes
df[['height_cm', 'weight_kg']] = df['Measurements'].str.split('/', expand=True)
```

```python
# Removing the strings and converting to numeric data type
df['height_cm'] = pd.to_numeric(df['height_cm'].str.strip(' cm'), errors='coerce')
df['weight_kg'] = pd.to_numeric(df['weight_kg'].str.strip(' kg'), errors='coerce')
```
